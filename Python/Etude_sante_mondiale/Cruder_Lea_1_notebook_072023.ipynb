<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personnes qui pourraient être nourries</h3>

Combien mange en moyenne un être humain ?

C'est une question très difficile à chiffrer au niveau mondial.

Pour les nutritionnistes, plusieurs paramètres rentrent en compte : la taille, l'âge, le poids, et le genre.
On calcule d'abord le metabolisme de base avec la formule de Black et al. :

Pour un homme : 1.083 x P0.48 x T0.50 x A-0.13
Pour une femme : 0.963 x P0.48 x T0.05 x A-0.13

Puis on le multiplie par un coefficient suivant l'activité sportive. 

Comme nous n'avons pas l'information de l'activité sportive de chacun, nous partons sur le coefficient moyen 1.63, qui correspond à une activité sportive minimum.
De manière générale, plus on est actif plus on a besoin d'apports.

Pour une population active sportivement, l'ANSES estime en moyenne un besoin énergétique de 2600 kcal/j pour les hommes et de 2100 kcal/j pour les femmes.

Ici on ne parle que des adultes actifs sportivement.


Source : AVIS et RAPPORTS de l'Anses relatifs à l'Actualisation des repères du PNNS 2017


D'après le rapport du projet de loi d'orientation agricole, disponible sur le site du Sénat, un être humain est considéré en sous-nutrition si son apport journalier est inférieur à 2.000 calories et à 10 grammes de protéines animales pour maintenir le poids corporel et permettre une activité, même légère.

La FAO présente également 2.000 kcal comme besoin journalier dans ses rapports, mais ajoute que pour une alimentation saine on sera plus à 2.330  kcal / jour.


Sources : Site du Sénat, Rapport du Projet de loi d'orientation agricole et site de la FAO


Nous tenons compte également qu'environ 65% de la population est située entre 15 à 64 ans (Source : INSEE).


Dans le cadre de cette étude, nous partirons sur le besoin au sens strict du terme avec une activité sportive légère,  et donc sur une moyenne de 2.000 calories par jour. (2350 pour les sportifs)


```python
# Préparation de la table dispo alimentaire en additionnant les différents éléments par pays
# afin d'ajouter dans ce dernier la population


dispo_alimentaire_totale = dispo_alimentaire.groupby('Zone')[['Aliments pour animaux (en Kg)', 'Autres Utilisations (en Kg)',
                                                                'Disponibilité alimentaire (Kcal/personne/jour)', 
                                                                'Disponibilité alimentaire en quantité (kg/personne/an)',
                                                                'Disponibilité de matière grasse en quantité (g/personne/jour)',
                                                                'Disponibilité de protéines en quantité (g/personne/jour)',  
                                                                'Disponibilité intérieure (en Kg)', 'Exportations - Quantité (en Kg)',
                                                                'Importations - Quantité (en Kg)', 'Nourriture (en Kg)',
                                                                'Pertes (en Kg)', 'Production (en Kg)', 'Semences (en Kg)',
                                                                'Traitement (en Kg)', 'Variation de stock (en Kg)'                                                       
                                                              ]].sum()
dispo_alimentaire_totale = dispo_alimentaire_totale.reset_index()
```

```python
# Jointure

dispo_alimentaire_totale = dispo_alimentaire_totale.merge(population_2017, on='Zone') 
```

```python
# Affichage du nouveau dataframe

dispo_alimentaire_totale.head(10)
```

```python
# Affichage des infos du nouveau dataframe :

dispo_alimentaire_totale.info
```

```python
# Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement

dispo_alimentaire_totale['dispo_kcal']=((dispo_alimentaire_totale['Disponibilité alimentaire (Kcal/personne/jour)'])*365)*dispo_alimentaire_totale['Population']
```

```python
# Affichage des 5 premières lignes

dispo_alimentaire_totale.head()
```

```python
# Calcul de la dispo alimentaire en kcal mondiale

dispo_kcal_mondiale = int(sum(dispo_alimentaire_totale['dispo_kcal'])/1000000)

print("La disponibilité alimentaire mondiale est de {:,} gigacalories.".format(dispo_kcal_mondiale))      
```

```python
# Calcul du nombre d'humains pouvant être nourris

nb_personnes_pouvant_etre_nourries =  int(sum(dispo_alimentaire_totale['dispo_kcal'])/2000/365)
print("Avec la disponibilité mondiale de Kcal, on pourrait nourrir", "{:,}".format(nb_personnes_pouvant_etre_nourries), 
      "d'êtres humains par jour.")
```

```python
# Calcul du nombre d'humains sportifs pouvant être nourris

nb_personnes_pouvant_etre_nourries =  int(sum(dispo_alimentaire_totale['dispo_kcal'])/2350/365)
print("Avec la disponibilité mondiale de Kcal, on pourrait nourrir", "{:,}".format(nb_personnes_pouvant_etre_nourries), 
      "par jour d'êtres humains sportivement actifs.")
```

Conclusion : Même si toute la population était active sportivement, la disponibilité mondiale en Kcal serait suffisante pour nourrir tout le monde.

```python
# Analyse des proportions

quantite_produite_noix =  int(chiffres_noix_bangladesh['Production (en Kg)'])
quantite_importee_noix = int(chiffres_noix_bangladesh['Importations - Quantité (en Kg)'])
dispo_interieure_noix = int(chiffres_noix_bangladesh['Disponibilité intérieure (en Kg)'])
quantite_exportee_noix = int(chiffres_noix_bangladesh['Exportations - Quantité (en Kg)'])

part_exportation_noix = round((quantite_exportee_noix/quantite_produite_noix)*100)
part_importation_noix = round((quantite_importee_noix/dispo_interieure_noix)*100)


print("D'après nos données, le Bangladesh exporte {}% de sa production en noix.\n".format(part_exportation_noix),
        "Il importe pourtant {}% de sa dispo totale.".format(part_importation_noix))
```
